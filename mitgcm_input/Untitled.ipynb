```python
import xarray as xr
```

```python
ds = xr.open_dataset('bcg_tau_obs_3.nc')

for var in ds.variables:
    newvar = 'belli_' + var.split('_')[1] 
    ds = ds.rename({var: newvar})

ds.to_netcdf('bcg_tau_obs_3b.nc')
```

```python
ds = xr.open_dataset('bcg-grid.nc')
ds
```

```python
for var in ds.variables:
    if var.split('_')[0]!='ihop':
        continue
    if var.split('_')[1]=='ranges' or var.split('_')[1]=='xc' or var.split('_')[1]=='yc':
        newvar = 'belli_' + var.split('_')[1]
        ds = ds.rename({var: newvar})
    elif var.split('_')[1]=='idw':
        newvar = 'belli_' + var.split('_')[1] + '_' + var.split('_')[2]
        ds = ds.rename({var: newvar})

ds.to_netcdf('bcg_gridb.nc')
```
